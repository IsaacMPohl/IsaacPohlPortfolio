```python
#finding how many unique departements their are and how many purchasers where made per apartment
uniqueDepartement=df['department'].unique()
my_dictionary = {}

#appending them to dictionary with value of 0
for dep in uniqueDepartement:
    my_dictionary[f'{dep}'] = 0


#going through each department 
for i,department in enumerate(df['department']):
    #if the department is in the dictionary/unqiue
    if department in my_dictionary:
        #add the cost of that row to the dicitonary
        my_dictionary[department]=my_dictionary[department]+df['cost'][i]

print(my_dictionary)
#opening up a new file of answers and writing to it
cleanFile = open("CleanData/Answers.csv","w",encoding="utf8")
cleanFile.write("department,cost\n")
for dep in my_dictionary:
    cleanFile.write(f"{dep},{my_dictionary[dep]} \n")
    
cleanFile.close()
```

```python
#opening up the file of answers to graph it
df = pd.read_csv("CleanData/Answers.csv",header=0)

largestNumber = 22e5

#going through each line of the answer file
for i,department in enumerate(df['department']):
    #if the cost is greater then 22e5 graph
    if int(df['cost'][i])>largestNumber:
        totals=df[df['department']==department].groupby('department')['cost']
        totalSum = totals.sum()
        totalKeys = totalSum.keys()
        plt.bar(totalKeys,totalSum,label=int(totalSum))
        
#adding graph information
plt.ylabel("Cost")
plt.xlabel("Department")
plt.title("Top 5 departments")
plt.ylim(2e3)
plt.legend(bbox_to_anchor=(1.1,1.07))
plt.show()
```
